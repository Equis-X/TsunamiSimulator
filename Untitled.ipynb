## Introduccin:
Este programa tiene como finalidad el entender el comportamiento de las secuencias de olas de gran tamao movidas por el desplazamiento de largos cuerpos de agua como es el tsunami, generalmente causados por terremotos, volcanes sub acuticos y otros fenmenos, una vez generado, el tsunami se propaga radialmente hacia el exterior de la superficie ocenica y asimismo crece hacia alturas bastante peligrosas mientras ms se acerca a aguas menos profundas.

### Archivo 1:
---

Primero que nada, vamos a colocar las importaciones necesarias para que pueda funcionar bien nuestro generador de datos (esto para poner a prueba el uso de las distintas ecuaciones, evitando el tener que estar conectado en lnea para su uso).
En este caso solo sern las libreras sys y numpy, numpy al ser la librera ms completa que conozco de cara a anlisis de datos y sys o sistema, solamente usada para poder colocar las salidas de sistema, con la finalidad de que no se inicien en caso de que los valores ingresados no cumplan con dadas caractersticas que veremos a continuacin.

````python
import sys
import numpy as np
````

Luego vamos con las declaraciones iniciales:

espacio, tiempo

dimensiones que se usaran ms adelante para bucles, el ndice central

tiempo por paso, tamao de la rejilla, velocidad de flujo de fondo, decadencia

````python
i, n = 0, 0
grid_size, num_time_steps, icenter = 100, 100, 25
dt, dx, c, decay = 1., 1., 1., 0.02
h, dh = [0] * grid_size, [0] * grid_size
````

Luego, revisaremos las condiciones que haran que este ejercicio no se pueda hacer a causa de las bases que plantean las ecuaciones, y que en caso de que se den, pues se har uso de sys y se saldr de ejecucin, lanzando determinada excepcin.

````python
if (grid_size <= 0): sys.exit('grid_size must be > 0')
if (dt <= 0): sys.exit('time step dt must be > 0')
if (dx <= 0): sys.exit('grid spacing dx must be > 0')
if (c <= 0): sys.exit('background flow speed c must be > 0')
````

Procederemos a dar ahora un valor a cada rejilla de la lista h, hecha para almacenar las diversas alturas que va a tener la ola. Tenemos que sea un bucle que se realice hasta que se llegue al tamao especificado en grid_size. Luego de tal, se har el clculo del exponencial de la decadencia al negativo por el valor del iterador menos el ndice inicial al cuadrado. Esto dndonos valores diferentes cada vez en forma de ola, iniciando desde abajo, llegando a un pico y bajando de nuevo.

Aqu veremos cual es el resultado que nos imprime este procedimiento tambin, formateado de una forma especfica la cual har ms fcil el trabajo de lectura del segundo archivo (a causa de que se capturara la salida de este para ejecutar el segundo).

````python
while (i != grid_size):
    h[i] = np.exp(-decay * (i - icenter)**2)
    i += 1

print(*h, sep="  ")
````

Ahora, se proceder a hacer lo mismo con el resto de los valores el nmero de veces especificado en la variable num_time_steps.
Este va a fluir teniendo en cuenta el flujo de fondo, o flujo trasero, siendo este el flujo que sigue a la ola en su movimiento. En este caso, vemos en el segundo bucle dentro del bucle principal el uso de una de las ecuaciones mencionadas al inicio:


NOTA: Los datos generados seran mostrados, son una cantidad de 100 * 100 datos, y para fines de demostracion prefiero no dejarlos fuera del documento.

````python
while n != num_time_steps:
    dh[0] = h[0] - h[grid_size - 1]

    i = 1
    while i != grid_size:
        dh[i] = h[i] - h[i - 1]
        i += 1

    i = 0
    while i != grid_size:
        h[i] = h[i] - c * dh[i] / dx * dt
        i += 1

    print(*h, sep="  ")
    n += 1
````

---

Con esto finalizamos el primer archivo Python y tenemos los datos generados.

Pero hey, como vamos a recolectar los datos en un archivo si nunca llevamos la salida a un archivo.
Esto para mi facilidad lo logre capturando la salida y almacenndola en un archivo. Esto convirtiendo el cdigo que tenamos anteriormente en lo siguiente (ojo a los comentarios, demarcan a los cambios agregados).

````python
import sys
import numpy as np

i, n = 0, 0
grid_size, num_time_steps, icenter = 100, 100, 25
dt, dx, c, decay = 1., 1., 1., 0.02
h, dh = [0] * grid_size, [0] * grid_size

if (grid_size <= 0): sys.exit('grid_size must be > 0')
if (dt <= 0): sys.exit('time step dt must be > 0')
if (dx <= 0): sys.exit('grid spacing dx must be > 0')
if (c <= 0): sys.exit('background flow speed c must be > 0')

while (i != grid_size):
    h[i] = np.exp(-decay * (i - icenter)**2)
    i += 1

# NUEVO
original_stdout = sys.stdout

# NUEVO
with open('1.txt', 'w') as f:
    # NUEVO
    sys.stdout = f
    print(*h, sep="  ")

    while n != num_time_steps:
        dh[0] = h[0] - h[grid_size - 1]

        i = 1
        while i != grid_size:
            dh[i] = h[i] - h[i - 1]
            i += 1

        i = 0
        while i != grid_size:
            h[i] = h[i] - c * dh[i] / dx * dt
            i += 1

        print(*h, sep="  ")
        n += 1
    # NUEVO
    sys.stdout = original_stdout
````

Esto nos presenta un nuevo uso de sys, en el cual podemos redireccionar hacia donde se va la salida que mostramos por la consola, llevando esta misma a un archivo f el cual es 1.txt, nuestro archivo dedicado al almacenamiento de los valores de salida para el procesamiento.

Oficialmente con esto, terminamos el archivo 1.
